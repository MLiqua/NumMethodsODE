```julia
function second_heunRKZ(dt, t_end, y0)
    t = 0:dt:t_end
    y = zeros(length(t))
    y[1] = y0
    
        for i in 1:length(t)-1 
            localx = t[i]
            localy = y[i]
                k1 = 3.0 * exp(-localx) -.4(localy)
            localx = t[i] + dt
            localy = y[i] + k1*dt
                k2 = 3.0 * exp(-localx) -.4(localy)
    
            y[i+1] = y[i] + (.5*k1 + .5*k2)*dt

        end
        return t, y
end
```

```julia
t_end = 3.0
y0 = 5.0
dt = 1.5

t_KRZ, y_KRZ = second_heunRKZ(dt, t_end, y0)
;
```

```julia
using Plots
plot(t_KRZ, y_KRZ, label= "Solution", lw =2)
xlabel!("time")
ylabel!("y(t)")
title!("RKZ Heun's Method")
```
