```julia
#=
Solving:
dy/dx = 3e^(-x) - .4y

Inital condition: y(0) = 5
Desired value: y(3.0) = ?
Step size (h) = 1.5
=#
```

```julia
function fourth_heunRKZ(dt, t_end, y0)
    t = 0:dt:t_end
    y = zeros(length(t))
    y[1] = y0
    
        for i in 1:length(t)-1 
            localx = t[i]
            localy = y[i]
                k1 = 3.0 * exp(-localx) -.4(localy)
            localx = t[i] + 1/2*dt
            localy = y[i] + 1/2*k1*dt
                k2 = 3.0 * exp(-localx) -.4(localy)
            localx = t[i] + 1/2*dt
            localy = y[i] + 1/2*k2*dt
                k3 = 3.0 * exp(-localx) -.4(localy)
            localx = t[i] + dt
            localy = y[i] + k3*dt
                k4 = 3.0 * exp(-localx) -.4(localy)
        
            y[i+1] = y[i] + 1/6*dt*(k1 + 2*k2 + 2*k3 + k4)

        end
        return t, y
end
```

```julia
t_end = 3.0
y0 = 5.0
dt = 1.5

t_KRZ, y_KRZ = fourth_heunRKZ(dt, t_end, y0)
```

```julia
using Plots
plot(t_KRZ, y_KRZ, label= "Solution", lw =2)
xlabel!("time")
ylabel!("y(t)")
title!("4th Order RKZ Runge Method")
```
