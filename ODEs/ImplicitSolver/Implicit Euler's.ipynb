```julia
function implicitEuler(dt, tEnd, y0, x0)
    t = 0.0:dt:tEnd
    y = zeros(length(t))
    x = zeros(length(t))
    y[1] = y0
    x[1] = x0
    
    
    for i in 1:length(t)-1
        x[i+1] = x[i] + dt
        y[i+1] = (y[i] + x[i+1]*dt) /(1.0 - (1.0/10.0)*(x[i+1])*dt)
    end
    return t, y
end
```

```julia
y0 = 1
x0 = 0
dt = 1
tEnd = 3

tVal, yplus1Val = implicitEuler(dt,tEnd,y0,x0)
```

```julia
using Plots
plot(tVal, yplus1Val, title = "Impliciy Eulers", xlabel = "Time", ylabel = "Y Plus I Value")
```
